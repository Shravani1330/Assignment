```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data from your CSV file
csv_file = 'Minor_project_dataset_CSE-AIML.csv'
df = pd.read_csv(csv_file)

# Access the "CGPA OVERALL" and "ATTENDANCE PERCENTAGE" columns
cgpa_data = df['CGPA OVERALL:']
attendance_data = df['ATTENDANCE PERCENTAGE:']

# Define the clustering criteria
criteria = []

for cgpa, attendance in zip(cgpa_data, attendance_data):
    if cgpa >= 80 and attendance >= 80:
        criteria.append("High CGPA & High Attendance")
    elif cgpa >= 80 and attendance < 80:
        criteria.append("High CGPA & Low Attendance")
    elif cgpa < 80 and attendance < 80:
        criteria.append("Low CGPA & Low Attendance")
    else:
        criteria.append("Low CGPA & High Attendance")

# Define colors for each cluster
color_dict = {
    "High CGPA & High Attendance": 'b',
    "High CGPA & Low Attendance": 'g',
    "Low CGPA & Low Attendance": 'r',
    "Low CGPA & High Attendance": 'c'
}

# Create a clustering graph with different colors based on the clustering criteria
plt.figure(figsize=(8, 6))
for cluster, color in color_dict.items():
    indices = [i for i, c in enumerate(criteria) if c == cluster]
    plt.scatter(cgpa_data[indices], attendance_data[indices], c=color, label=cluster)

plt.xlabel('CGPA OVERALL:')
plt.ylabel('ATTENDANCE PERCENTAGE:')
plt.title('Clustering Plot for CGPA and Attendance')
plt.legend()
plt.show()
```
